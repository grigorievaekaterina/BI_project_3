# K-mer profile and genome size estimation

Let us count k-mers

We are going to use k-mer sizes of 31, count kmers with *jellyfish count* and make a histogram file with *jellyfish histo*

```python
!jellyfish count -t 8 -m 31 -C -s 50000000 -o k31mer "/content/SRR292678sub_S1_L001_R1_001.fastq" 
```

```python
!jellyfish histo -o k31mer.histo k31mer
```

Then we need R to visualize the histo

```python
!pip install rpy2==3.4.2
```

```python
%load_ext rpy2.ipython
```

```python
%%R 
k31mer_histo <- read.table("/content/k31mer.histo")
plot(k31mer_histo[5:200,],type="l")
```

Then we want to know the maximum value (the peak)

```python
%%R
k31mer_histo[50:70,]
```

Max value is 62. Let us estimate the genome size of our bacteria with the following formulas:

N = (M*L)/(L-K+1)

Genome_size = T/N

(N: Depth of coverage, M: Kmer peak, K: Kmer-size, L: avg read length T: Total bases)

```python
N = (62*150)/(150-31+1)
T = 150*5499346
Genome_size = T/N
Genome_size
```

So, genome size is 10643895 bp.

# Tracing the source of toxin genes in E. coli X

```python
n=39
arr = []
with open('scaffolds.gbk', 'r') as file:
  while len(arr) < n:
    line = file.readline()
    if 'product' in line:
      arr.append(line)
  for line in file:
    if 'product' in line:
      arr = arr[1:]
      arr.append(line)
    if 'Shiga' in line:
      break
  while len(arr) < 2*n:
    line = file.readline()
    if 'product' in line:
      arr.append(line)
```

```python
for element in arr:
  if not 'hypothetical' in element:
    print(element.split('\"')[1])
```
